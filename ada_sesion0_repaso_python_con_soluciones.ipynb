### 2. Medición de tiempo de ejecución

A la hora de medir el tiempo de ejecución de un código fuente, es esencial diferenciar dos términos:

- El **wall time** es el tiempo total transcurrido durante la medición y que por tanto podríamos medir con un cronómetro. Es la diferencia entre el tiempo en el que un programa terminó su ejecución y el tiempo en el que el programa comenzó. Pero es importante indica que dicho tiempo también **incluye el tiempo de espera para obtener recursos**.

- El **CPU time** se refiere al tiempo que la CPU estuvo ocupada procesando las instrucciones del programa. Sin embargo, el tiempo empleado en esperar a que se completen otras tareas (como las operaciones de E/S) no se incluye en el tiempo de CPU asi como tampoco el tiempo de espera de recursos.

La diferencia entre el *wall* y el *CPU time* puede deberse a la arquitectura de la máquina o a dependencias en tiempo real como, por ejemplo, a retrasos programados o a la espera de que los recursos del sistema estén disponibles.

Por ejemplo, si para un programa obtenemos un *CPU time* de 0m0.2s y un *wall time* de 2m4s, significa que el programa estuvo activo durante 2 minutos y cuatro segundos. Sin embargo, el procesador del ordenador sólo empleó 0,2 segundos realizando cálculos para el programa.




En el entorno de ejecución con `Notebooks` podemos usar la primitiva `%time` que nos devolverá ambos tipos de tiempo como salida de la celda en la que se encuentre.

```python
%time

n=0
for i in range(100000):
  n += i
```

En otros entornos de ejecución, el módulo `time` proporciona funciones para trabajar con el tiempo.

- Para medir el *wall time* podemos usar la función `.time`
- Para medir el *cpu time* podemos usar la funcion `.process_time`


Así, para medir ambos tiempos de un determinado código solo se tiene que llamar a ambas funciones antes y después de la funcionalidad a testear.

```python
import time

start_time = time.time()
start_time_perf= time.process_time()
# Código que queremos medir.
n=0
for i in range(100000):
  n += i
time.sleep(3) #Entramos a dormir para forzar que el wall time sea claramente mayor al cpu time.
# Fin del código que queremos medir
end_time_time_perf= time.process_time()
end_time = time.time()

cpu_time= end_time_time_perf - start_time_perf
wall_time = end_time - start_time
print("CPU time:", cpu_time, "segundos")
print("Wall time:", wall_time, "segundos")
```

### 3. Paso de parámetros en Python


Dependiendo del tipo de dato que enviemos a la función, podemos diferenciar dos comportamientos:

- **Paso por valor**: Se crea una copia local de la variable dentro de la función.
- **Paso por referencia**: Se maneja directamente la variable, los cambios realizados dentro de la función le afectarán también fuera.
Tradicionalmente:

Los tipos simples se pasan por valor: Enteros, flotantes, cadenas, lógicos...

Los tipos compuestos se pasan por referencia: Listas, diccionarios, conjuntos...


#### Paso por valor

```python
def doblar_valor(n):
    n *= 2
    print(n)

n = 10
doblar_valor(n)
print(n)
```

Como vemos el valor de `n` no se modifica después de la llamada a la función.

#### Paso por referencia

```python
def doblar_valores(numeros):
    for i,n in enumerate(numeros):
        numeros[i] *= 2

ns = [10,50,100]
doblar_valores(ns)
print(ns)
```

En este caso vemos que la lista `ns` sí que se modifica al hacer la llamada a la función `doblar_valores`.

### 4. Recursividad

**¿Qué es la recursividad?**

La recursividad es un concepto de programación en el que una función se llama a sí misma para resolver un problema de manera iterativa.

Es especialmente útil cuando el problema se puede dividir en casos más pequeños y similares al original.

Un ejemplo muy sencillo de aplicación de la recursividad consiste en el cómputo del factorial de un número.

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)

print(factorial(5))
```

Si añadimos un `print` al comienzo de la función podemos ver la secuencia encadenada de llamadas.

```python
def factorial(n):
  print("Procesando el número ",n)
  if n == 0:
    print("Entro en el caso base")
    return 1
  else:
    return n * factorial(n - 1)

print("El valor del factorial es", factorial(5))
```

A la hora de usar la recursividad debemos de tener en cuenta cómo se realiza el paso de parámetros a una función en Python que hemos visto en el punto 3.

Por ejemplo, al pasarse las listas por referencia, podríamos implementar una función recursiva que vaya generando una lista incremental simplememente añadiendo un nuevo elemento que sea el último de la lista actual más  `1`.

```python
def añadir_elemento(l):
  if len(l)==5:
    return l
  else:
    l.append(l[-1]+1)
    return añadir_elemento(l)

mi_lista=[1]
nueva_lista= añadir_elemento(mi_lista)
print(mi_lista)
```

### 5. Generación de `Docstrings`

Cuando deseemos documentar nuestro código, ya sea una función, un método, una clase, un módulo o un paquete, lo haremos mediante un `docstring`.

Para poder documentar nuestro código vamos a colocar, en la primera línea del cuerpo de la función, un comentario con triple comillas simples (""") con los siguientes puntos:

- La primera línea debe poseer una breve descripción del funcionamiento del código. Las descripciones deben comenzar con mayúscula y finalizar con punto.

- La segunda deberá ser un espacio en blanco, de tal forma que sea fácil distinguir entre la descripción y el resto del comentario.

- La tercera línea deberá incluir una descripción más detallada de la funcionalidad del código, seguida por un espacio en blanco.

- La cuarta línea deberá ser de nuevo un espacio en blanco.

- Las siguientes líneas deben ser uno o más párrafos que describan lps parámetros de entrada de la función.

- La última línda deberá ser el tipo de dato devuelvo por la función y una breve descripción del resultado.

A continuación mostramos un ejemplo de `docstring` para una función sencilla que trata una función. ¿Sabrías decir lo que hace sin necesidad de ejecutar el código?

```python
def sustituir_entre_indices(mi_cadena, viejo_caracter, nuevo_caracter, i, j):
    """
    Sustituye todas las ocurrencias del caracter 'viejo_caracter' por el caracter 'nuevo_caracter' en la cadena 'mi_cadena' entre los índices dados.

    Esta función toma una cadena de texto mi_cadena, dos caracteres 'viejo_caracter' y 'nuevo_caracter', y dos índices 'i' y 'j', y reemplaza
    todas las ocurrencias del caracter 'viejo_caracter' por el caracter 'nuevo_caracter' en la subcadena de mi_cadena delimitada por los índices 'i' y 'j'.

    Parámetros:
    mi_cadena (str): La cadena de texto en la que se realizará la sustitución.
    viejo_caracter (str): El caracter que se desea reemplazar.
    nuevo_caracter (str): El caracter con el que se reemplazará 'a'.
    i (int): El índice de inicio (inclusive) del rango donde se realizará la sustitución.
    j (int): El índice de fin (exclusivo) del rango donde se realizará la sustitución.

    Devuelve:
    str: La cadena 'mi_cadena' con todas las ocurrencias de 'nuevo_caracter' reemplazadas por 'b' entre los índices 'i' y 'j'.
    """
    parte_inicial= mi_cadena[:i]
    parte_reemplazada= mi_cadena[i:j].replace(viejo_caracter, nuevo_caracter)
    parte_final= mi_cadena[j:]
    return parte_inicial + parte_reemplazada + parte_final
```

Vamos a ejecutar dicha función para ver si se corresponde con lo que has deducido de leer su `docstring`.

```python
cadena_original = "Hola mundo"
caracter_a = "o"
caracter_b = "x"
indice_inicio = 1
indice_fin = 5
resultado_sustitucion = sustituir_entre_indices(cadena_original, caracter_a, caracter_b, indice_inicio, indice_fin)
print("Resultado de la sustitución entre los índices", indice_inicio, "y", indice_fin, ":", resultado_sustitucion)
```

**Imágenes en 3D**

Para utilizar gráficos 3D en matplotlib, primero tenemos que crear una instancia de la clase `Axes3D`. Los ejes 3D se pueden añadir a un lienzo de matplotlib exactamente de la misma manera que los ejes 2D; o, más convenientemente, pasando un argumento `projection='3d'` a los métodos `add_axes` o `add_subplot`.

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5
def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T

fig = plt.figure(figsize=(14,6))

# `ax` is a 3D-aware axis instance because of the projection='3d' keyword argument to add_subplot
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# surface plot con diferentes tipos de texturas.
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```

Implementa una función recursiva que genere una lista de longitud 5 en donde todos sus elementos sean el número 1, es decir [1,1,1,1,1].

No te olvides de generar el `docstring` de dicha función.

Solución

```python
def añadir_elemento(l):
  """
  Función que genera una lista con 5 números 1.

  La función hace uso de la recursividad para componer una lista 'l' paso a paso añadiendo elementos al final de la lista

  Parámetros:
  l (list): Lista inicial vacía.

  Devuelve:
  array: lista de 5 elementos en donde todos los elementos son el número 1.

  """
  if len(l)==5:
    return l
  else:
    l.append(1)
    return añadir_elemento(l)

mi_lista=[]
nueva_lista= añadir_elemento(mi_lista)
print(mi_lista)
```
